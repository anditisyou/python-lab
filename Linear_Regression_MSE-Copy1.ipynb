```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dp = pd.read_csv("Housing.csv")
X = dp['price'].values   # column price
Y = dp['area'].values    # column area

#show all training examples
for i in range(len(X)):
    #print("(X[",i,"],Y[",i,"]) =",X[i],Y[i])
    print(f"(X[i],Y[i]) = ({X[i]},{Y[i]})")
    
```

```python
# Scatter plot 
plt.scatter(X,Y,marker="x", c='r', label = "input")
plt.xlabel("inputs")
plt.ylabel("outputs")
plt.legend()
plt.show()
```

```python
#linear regression model y = wx + b
def linear_regression_model(X,w,b):
    f_x = np.zeros(len(X))
    for i in range(len(X)):
        f_x[i] = w * X[i] + b
    return f_x

# initializing the w and b for the regression
w=0.0009
b=1000

# printing the output from the linear regression model
l_r=linear_regression_model(X, w, b)
print(l_r)
```

```python
#plot linear regression model
plt.plot(X,l_r, color = "skyblue")
plt.scatter(X,Y,marker="x", c='r', label = "input")
plt.xlabel("inputs")
plt.ylabel("outputs")
plt.legend()
plt.show()
```

## ***COST FUNCTION***

```python
def cost_function_linear_regression(pred_y,Y,m):
    sum_cost=0
    
    for i in range(m):
        #pred_y=w*X[i]+b
        sum_cost=sum_cost+(pred_y[i]-Y[i])**2
    cost_total=(1/2*m)*sum_cost
    return cost_total
```

```python
pred_y=linear_regression_model(X,w,b)
m=len(X)
print(cost_function_linear_regression(pred_y,Y,m))
```

## ***Gradient descents***

```python
#Gradient descent
def gradient_descent_cost_function(X,Y,w,b):
    j_dw = 0 
    j_db = 0
    for i in range(m):
        f_x = w * X[i] + b
        j_dw = j_dw + (f_x - Y[i]) * X[i]
        j_db = j_db + (f_x - Y[i])

    j_dw = j_dw / m
    j_db = j_db / m
    return j_dw,j_db
```

```python
j_dw,j_db = gradient_descent_cost_function(X,Y,w,b)
print(f"j_dw = {j_dw} and j_db = {j_db}")
```

```python
#Gradient descent algorithm
def gradient_descent_algorithm(X,Y,w,b,learning_rate,gradient_descent_cost_function,iteration):
    w = w
    b = b
    for i in range(iteration):
        j_dw,j_db = gradient_descent_cost_function(X,Y,w,b)
        w = w - learning_rate * j_dw
        b = b - learning_rate * j_db
    return w,b
```

```python
w = 0
b = 0
learning_rate = 0.0001 # Best range for the learning rate is 0.001 to 0.0001, where 0.1 to 0.00001 is the actual rate range
iteration = 24
w,b = gradient_descent_algorithm(X,Y,w,b,learning_rate,gradient_descent_cost_function,iteration)
```

```python
print(f"w = {w} and b = {b}")
l_r=linear_regression_model(X, w, b)
```
