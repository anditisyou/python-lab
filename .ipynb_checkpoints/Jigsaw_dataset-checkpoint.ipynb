```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("jigsaw-toxic-comment-train.csv")
categories = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
counts = {cat: df[cat].value_counts() for cat in categories}
counts_df = pd.DataFrame(counts).T
counts_df.columns = ['Not Present (0)', 'Present (1)']
print(counts_df)
counts_df.plot(kind='bar', figsize=(10,6))
plt.title("Distribution of Toxicity Labels")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Label")
plt.tight_layout()
plt.show()
```

```python
df['toxic_obscene'] = df[['toxic','obscene']].max(axis=1)
df['identity_insult'] = df[['insult','identity_hate']].max(axis=1)
df['severe_threat'] = df[['threat','severe_toxic']].max(axis=1)
df.to_csv("merged_file.csv", index=False)
```

```python
df = pd.read_csv("merged_file.csv")
df.drop(['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate'], axis=1)
categories = ['toxic_obscene', 'identity_insult', 'severe_threat']
counts = {cat: df[cat].value_counts() for cat in categories}
counts_df = pd.DataFrame(counts).T   # transpose so categories are rows
counts_df.columns = ['Not Present (0)', 'Present (1)']
print(counts_df)
counts_df.plot(kind='bar', figsize=(10,6))
plt.title("Distribution of Toxicity Labels")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Label")
plt.tight_layout()
plt.show()
```
