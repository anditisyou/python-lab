```python
import pandas as pd
import numpy as np
input_data=pd.read_csv("similarity (1).csv")
input_data
l1 = input_data.values
l1
```

```python
def manhattan_distance_function(x, y): 
    return np.sum(abs(x-y))
def euclidean_distance_function(x,y):
    return np.sqrt(np.sum((x-y)**2))
def cousine_distance_function(x,y):
    return np.dot(x,y)/(np.linalg.norm(x)*(np.linalg.norm(y)))
def minkowski_distance_function(x,y,p):
    n=1/p
    return (np.sum((x-y)**p))**n
```

```python
for i in range(len(l1)):
    for j in range(i+1,len(l1)):
        print("Euclidean distance ",euclidean_distance_function(l1[i],l1[j]))
for i in range(len(l1)):
    for j in range(i+1,len(l1)):
        print("manhattan_distance_function ",manhattan_distance_function(l1[i],l1[j]))
for i in range(len(l1)):
    for j in range(i+1,len(l1)):
        print("cousine_distance_function ",cousine_distance_function(l1[i],l1[j]))
```

```python

for i in range(len(l1)):
    for j in range(i+1,len(l1)):
        print(" minkowski_distance_function ", minkowski_distance_function(l1[i],l1[j],4))
```
