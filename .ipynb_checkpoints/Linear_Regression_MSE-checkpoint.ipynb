```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
X=np.array([1,2,3,4,5])
Y=np.array([200,300,400,500,600])

#show all training examples
for i in range(len(X)):
    #print("(X[",i,"],Y[",i,"]) =",X[i],Y[i])
    print(f"(X[i],Y[i]) = ({X[i]},{Y[i]})")
    
```

```python
# Scatter plot 
plt.scatter(X,Y,marker="x", c='r', label = "input")
plt.xlabel("inputs")
plt.ylabel("outputs")
plt.legend()
plt.show()
```

```python
#linear regression model y = wx + b
def linear_regression_model(X,w,b):
    f_x = np.zeros(len(X))
    for i in range(len(X)):
        f_x[i] = w * X[i] + b
    return f_x

#initilizing the w and b for the regression
w=100
b=100

#priting the output got from linear regression model
l_r=linear_regression_model(X, w, b)
print(l_r)
```

```python
#plot linear regression model
plt.plot(X,l_r, color = "skyblue")
plt.scatter(X,Y,marker="x", c='r', label = "input")
plt.xlabel("inputs")
plt.ylabel("outputs")
plt.legend()
plt.show()
```

```python
#Gradient descent
def gradient_descent_cost_function(X,Y,w,b):
    j_dw = 0 
    j_db = 0
    m = 5
    for i in range(m):
        f_x = w * X[i] + b
        j_dw = j_dw + (f_x - Y[i]) * X[i]
        j_db = j_db + (f_x - Y[i])

    j_dw = j_dw / m
    j_db = j_db / m
    return j_dw,j_db
```

```python
j_dw,j_db = gradient_descent_cost_function(X,Y,w,b)
print(f"j_dw = {j_dw} and j_db = {j_db}")
```

```python
#Gradient descent algorithm
def gradient_descent_algorithm(X,Y,w,b,learning_rate,gradient_descent_cost_function,iteration):
    w = w
    b = b
    for i in range(iteration):
        j_dw,j_db = gradient_descent_cost_function(X,Y,w,b)
        w = w - learning_rate * j_dw
        b = b - learning_rate * j_db
    return w,b
```

```python
w = 0 
b = 0
learning_rate = 0.001 # Best range for the learning rate is 0.001 to 0.0001 where 0.1 to 0.00001 actual rate range
iteration = 1000
w,b = gradient_descent_algorithm(X,Y,w,b,learning_rate,gradient_descent_cost_function,iteration)
```
